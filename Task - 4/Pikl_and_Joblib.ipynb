Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/hiring.csv")
df.head(10)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
import math
mn = math.floor(df['test_score(out of 10)'].mean())
mn
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(mn)
df
```

```python
!pip install word2number
```

```python
from word2number import w2n
df.experience = df.experience.fillna("zero")
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
df.rename(columns={'test_score(out of 10)': 'test_score', 'interview_score(out of 10)': 'interview_score', 'salary($)': 'salary'}, inplace=True)
df
```

```python
X = df[['experience', 'test_score', 'interview_score']]
y = df['salary']

reg = linear_model.LinearRegression()
reg.fit(X, y)
```

```python
prediction = reg.predict([[2, 9, 6]])[0]
print(f"Predicted salary for a candidate with 2 years experience, 9 test score, and 6 interview score: ${prediction:,.2f}")
```

```python
import joblib
joblib.dump(reg, 'model_hiring.pkl')

print("Linear Regression model saved as 'model_hiring.pkl'")
```
